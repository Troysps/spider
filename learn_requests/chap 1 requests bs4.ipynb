### BeautifulSoup重点：select 查询方法

```python
html = """
<html><head><title>The Dormouse's story</title></head>
   <body>
   <p class="title" name="dromouse"><b>The Dormouse's story</b></p>
   <p class="story">Once upon a time there were three little sisters; and their names were
   <a href="http://example.com/elsie" class="sister" id="link1"><!-- Elsie --></a>,
   <a href="http://example.com/lacie" class="sister" id="link2">Lacie</a> and
   <a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>;
   and they lived at the bottom of a well.</p>
   <p class="story">...</p>
   """
```

#### part 1 通过标签名查找

```python
from bs4 import BeautifulSoup
soup1 = BeautifulSoup(html, "lxml") # requests res.text   'html.paraser'
print(soup1.select('title'))
print(soup1.select('a'))
print(soup1.select('b'))
```

#### part 2 通过类名查找

```python
# class '.xxxxx'
print(soup1.select('.title'))
print(soup1.select('.story'))
```

#### part 3 通过id查找

```python
# id '#xxxxx'
print(soup1.select('#link1'))
print(soup1.select('#link2'))
print(soup1.select('#link3'))
```

#### part 4 组合查询

```python
# 将标签名 类名 id组合查询
print(soup1.select('p #link3'))
# 子标签查询
print(soup1.select('head > title'))
print(soup1.select('body > p'))
```

#### part 5 属性查找

```python
print(soup1.select('a[href="http://example.com/elsie"]'))
# 属性查找同样也可以通过组合查询
print(soup1.select('p a[href="http://example.com/elsie"]'))
```
