```python
# -*- coding:utf-8 -*-
import requests
from bs4 import BeautifulSoup

urls = ['https://www.taptap.com/category/Casual?page={}'.format(str(i)) for i in range(1,200)]
print(urls)
```

```python
def get_soup(url):
    res = requests.get(url)
    res.encoding = r'utf-8'
    soup = BeautifulSoup(res.text, 'html.parser')
    return soup
soups = []
for url in urls:
    soups.append(get_soup(url))
    
soups
```

```python
game_names, game_categories, game_overall_rating, game_link = [], [], [], []
def get_game_data(soup):
    for name in soup.select('h4.flex-text'):
        if name not in game_names:
            game_names.append(name.text.strip())
    for category in soup.select('a[href="https://www.taptap.com/category/casual"]'):
        if category not in game_categories:
            game_categories.append(category.text.strip())
    for rating in soup.select('.item-caption-label'):
        try:
            game_overall_rating.append(float(rating.contents[3].text.strip()))
        except IndexError:
            game_overall_rating.append(None)
    for link in soup.select('a[class="item-caption-title flex-text-overflow"]'):
        if link not in game_link:
            game_link.append(link['href'])

for soup in soups:
    get_game_data(soup)

```

```python
import numpy as np
import pandas as pd
data = {
    'game_names' : game_names,
    'game_categories' : game_categories,
    'game_link' : game_link,
    'game_overall_rating' : game_overall_rating
}

g_data = pd.DataFrame(data)
g_data
```
