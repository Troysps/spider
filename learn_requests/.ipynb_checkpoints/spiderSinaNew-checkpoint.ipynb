```python
# 完成内文信息抽取函数
# 建立字典将其进行存储
import requests
from bs4 import BeautifulSoup
import re
import json
from datetime import datetime
comment_url = 'http://comment5.news.sina.com.cn/page/info?version=1&format=js&channel=sh&newsid=comos-{}&group=&compress=0&ie=gbk&oe=gbk&page=1&page_size=20'
def getComment(url):
    m = re.search('doc-i(.*).shtml', url)
    news_id = m.group(1)
    comments = requests.get(comment_url.format(news_id))
    jd = json.loads(comments.text.strip('var = data'))
    return jd['result']['count']['total']
def getNewDetail(newsurl):
    result = {}
    res = requests.get(newsurl)
    res.encoding = u'utf-8'
    soup = BeautifulSoup(res.text, 'html.parser')
    try:
        result['title'] = soup.select('h1.main-title')[0].text
        result['news_source'] = soup.select('a.source')[0].text
        result['news_source'] = soup.select('a.source')[0].text
        timesource = soup.select('span.date')[0].text
        result['datetime'] = datetime.strptime(timesource, '%Y年%m月%d日 %H:%M')
        article = ''.join([p.text.strip() for p in soup.select('#article p')[:-1]])
        result['article'] = article.replace(u'\xa0', ' ')
        result['editor'] = soup.select('p.show_author')[0].text.strip()[5:]
        result['comment'] = getComment(newsurl)
    except IndexError:
        result['title'] = None
        result['news_source'] = None
        result['news_source'] = '来源不明'
        result['datetime'] = None
        result['article'] = None
        result['editor'] = None
        result['comment'] = getComment(newsurl)
    return result
getNewDetail('http://news.sina.com.cn/c/nd/2018-01-07/doc-ifyqkarr7761968.shtml')
```

```python
# 寻找分页信息 查找url
res = requests.get('http://api.roll.news.sina.com.cn/zt_list?channel=news&cat_1=gnxw&cat_2==gdxw1||=gatxw||=zs-pl||=mtjj&level==1||=2&show_ext=1&show_all=1&show_num=22&tag=1&format=json&page=2&callback=newsloadercallback&_=1515316474389')
res.encoding = u'utf-8'
res.text
jd = json.loads(res.text.lstrip('  newsloadercallback(').rstrip(');'))
jd['result']['data']
for ent in jd['result']['data']:
    print(ent['url'])
```

```python
def parserListLinks(url):
    newsdetail = []
    res = requests.get(url)
    res.encoding = u'utf-8'
    jd = json.loads(res.text.lstrip('  newsloadercallback(').rstrip(');'))
    for ent in jd['result']['data']:
        newsdetail.append(getNewDetail(ent['url']))
    return newsdetail
```

```python
url_1 = 'http://api.roll.news.sina.com.cn/zt_list?channel=news&cat_1=gnxw&cat_2==gdxw1||=gatxw||=zs-pl||=mtjj&level==1||=2&show_ext=1&show_all=1&show_num=22&tag=1&format=json&page=2&callback=newsloadercallback&_=1515316474389'
parserListLinks(url_1)
```

```python
url = 'http://api.roll.news.sina.com.cn/zt_list?channel=news&cat_1=gnxw&cat_2==gdxw1||=gatxw||=zs-pl||=mtjj&level==1||=2&show_ext=1&show_all=1&show_num=22&tag=1&format=json&page={}'
for i in range(1, 10):
    print(url.format(i))
```

```python
#　批量抓取
url_2 = 'http://api.roll.news.sina.com.cn/zt_list?channel=news&cat_1=gnxw&cat_2==gdxw1||=gatxw||=zs-pl||=mtjj&level==1||=2&show_ext=1&show_all=1&show_num=22&tag=1&format=json&page={}'
news_total = []
for i in range(1, 10):
    newsurl = url_2.format(i)
    newsary = parserListLinks(newsurl)
    news_total.extend(newsary)
```

```python
import pandas as pd
df = pd.DataFrame(news_total)
df
```
